#Experimental test of two models for the generation of oblique saccades

Implementation of van Gisbergen's models from the above named paper in Exp. Brain Res. 57 pp 321-336 (1985).

## Peak velocity vs. amplitude relation

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# R is the peak amplitude (degrees)
R=np.linspace(0,30,101)

# Rc represents and angular constant (degrees)
Rc=10

# VEC_a is the asymptotic vectorial peak velocity
VEC_a = 1000

# Eqn (1) from the paper
VEC = VEC_a*(1-np.exp(-R/Rc))

plt.plot(R, VEC)
```

## Peak velocity relation for components

```python
from mpl_toolkits.mplot3d import Axes3D

phi = np.linspace(0,0.5*3.14,101)
phi_, R_ = np.meshgrid(phi, R)

# Eqn (2) from the paper. Hp is horizontal peak vel.
Hp = VEC_a*np.cos(phi_)*(1-np.exp(-R_/Rc))

fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(phi_, R_, Hp, rstride=2, cstride=2, cmap=plt.cm.coolwarm,
        linewidth=0, antialiased=False)
ax.set_xlabel('phi')
ax.set_ylabel('R')
ax.set_zlabel('Hp')
```

```python
deltaH = R_ * np.cos(phi_)

# Eqn (3) from the paper
Hp = VEC_a*np.cos(phi_)*(1-np.exp(-deltaH/(Rc*np.cos(phi_))))

fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(phi_, R_, Hp, rstride=8, cstride=8, cmap=plt.cm.coolwarm,
        linewidth=0.3, antialiased=False)
ax.set_xlabel('phi')
ax.set_ylabel('deltaH')
ax.set_zlabel('Hp')
```

```python
deltaV = R_ * np.sin(phi_)

# Eqn (4) from the paper
Vp = VEC_a*np.sin(phi_)*(1-np.exp(-deltaV/(Rc*np.sin(phi_))))

fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(phi_, R_, Vp, rstride=8, cstride=8, cmap=plt.cm.coolwarm,
        linewidth=0.3, antialiased=False)
ax.set_xlabel('phi')
ax.set_ylabel('deltaV')
ax.set_zlabel('Vp')
```

##Fig 2 D

```python
Hp1 = R;

_phi = 0
Hp = VEC_a*np.cos(_phi)*(1-np.exp(-Hp1/(Rc*np.cos(_phi))))
plt.plot(R, Hp, label='1. phi=0')

_phi = np.pi/6
Hp = VEC_a*np.cos(_phi)*(1-np.exp(-Hp1/(Rc*np.cos(_phi))))
plt.plot(R, Hp, label='2. phi=30')

_phi = np.pi/3
Hp = VEC_a*np.cos(_phi)*(1-np.exp(-Hp1/(Rc*np.cos(_phi))))
plt.plot(R, Hp, label='3. phi=60')

plt.gca().set_xlabel('deltaH (deg)')
plt.gca().set_ylabel('Horiz peak vel. (deg/s)')
handles, labels = plt.gca().get_legend_handles_labels()
plt.gca().legend(handles, labels, loc=4)            
```
