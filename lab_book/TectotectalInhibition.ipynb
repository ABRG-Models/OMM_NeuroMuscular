This notebook is the development and demonstration of the tecto-tectal inhibitory python connection script for the Oculomotor model.

The input is a 2500 vector of x/y coordinates from a 50x50 grid.

The output is a list of connection weights between src/dst pairs.

```python
#PARNAME=sigma #LOC=1,1
#HASWEIGHT

def connectionFunc(srclocs,dstlocs,sigma):

	import math
	normterm = 1/(math.pow(sigma,2)*2*math.pi)

	i = 0
	out = []
	for srcloc in srclocs:
		j = 0
		for dstloc in dstlocs:
            
			s1 = srcloc[1]
			# Invert one dimension of the sources prior to dist calculation:
			s0 = -1*(srcloc[0]-(49.0/2.0))
			s2 = srcloc[2]
			d1 = dstloc[1]
			d0 = dstloc[0]-(49.0/2.0)
			d2 = dstloc[2]
            
			dist = math.sqrt(math.pow((s0 - d0),2)
                             + math.pow((s1 - d1),2)
                             + math.pow((s2 - d2),2))

			conn_weight = normterm*math.exp(-0.5*math.pow(dist/sigma,2))

			if conn_weight > 0.001:
				conn = (i,j,0,conn_weight)
				out.append(conn)

			j = j + 1

		i = i + 1

	return out
```

```python
srclocs = []
for i in range(0, 50):
    for j in range(0, 50):
        srcloc=[j,i,0]
        srclocs.append(srcloc)

result = connectionFunc (srclocs, srclocs, 4)

print "Done computing"

i = 0
for res in result:
    i = i + 1
    print res[0],res[1],res[2],res[3]
    if i>1000:
        break

#for srcloc in srclocs:
#    print srcloc[0],srcloc[1],srcloc[2]
```
